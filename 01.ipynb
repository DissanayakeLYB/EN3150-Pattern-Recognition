```python
import numpy as np
import matplotlib.pyplot as plt

def generate_signal(signal_length, num_nonzero):
    
    signal = np.zeros(signal_length)
    nonzero_indices = np.random.choice(signal_length, num_nonzero, replace=False)
    nonzero_values = 50*np.random.randn(num_nonzero)  
    signal[nonzero_indices] = nonzero_values
    return signal

signal_length = 100  # Total length of the signal
num_nonzero = 10    # Number of non-zero elements in the signal
your_index_no = 200145 # Enter without english letter and without leading zeros
signal = generate_signal(signal_length, num_nonzero)
signal[10] = (your_index_no % 10)*10 + 10

if  your_index_no % 10 == 0:
  signal[10] = np.random.randn(1) + 30


signal=signal.reshape(signal_length,1)
plt.figure(figsize=(15,5))
plt.subplot(1, 1, 1)
plt.title("Data")
plt.stem(signal)
```

```python
from sklearn.preprocessing import MaxAbsScaler

# MaxAbsScaler
max_abs_scaler = MaxAbsScaler()
max_abs_scaled_data = max_abs_scaler.fit_transform(signal)

# Min-Max Scaling
def min_max_scale(data):
    min_val = np.min(data)
    max_val = np.max(data)
    scaled_data = (data - min_val) / (max_val - min_val)
    return scaled_data

min_max_scaled_data = min_max_scale(signal)

# Standardization
mean = np.mean(signal)
std_dev = np.std(signal)
standardized_data = (signal - mean) / std_dev

```

```python
# Plot original and normalized data
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.title("Original Data")
plt.stem(signal)

plt.subplot(2, 2, 2)
plt.title("MaxAbsScaled Data")
plt.stem(max_abs_scaled_data)

plt.show()
```

```python
# Plot original and normalized data
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.title("Original Data")
plt.stem(signal)

plt.subplot(2, 2, 2)
plt.title("Min-Max Scaled Data")
plt.stem(min_max_scaled_data)

plt.show()
```

```python
# Plot original and normalized data
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.title("Original Data")
plt.stem(signal)

plt.subplot(2, 2, 2)
plt.title("Standardized Data")
plt.stem(standardized_data)

plt.show()


```

```python
# Count non-zero elements
num_nonzero_original = np.count_nonzero(signal)
num_nonzero_max_abs_scaled = np.count_nonzero(max_abs_scaled_data)
num_nonzero_min_max_scaled = np.count_nonzero(min_max_scaled_data)
num_nonzero_standardized = np.count_nonzero(standardized_data)

print("Non-zero elements in original data:", num_nonzero_original)
print("Non-zero elements after MaxAbsScaler:", num_nonzero_max_abs_scaled)
print("Non-zero elements after Min-Max Scaling:", num_nonzero_min_max_scaled)
print("Non-zero elements after Standardizing:", num_nonzero_standardized)
```
